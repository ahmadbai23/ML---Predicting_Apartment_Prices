### **Metric Evaluation**

Metrik evaluasi yang digunakan untuk mengitung performance dari machine learning yang digunakan yaitu **RMSE, MAE, dan MAPE**. Pada dataset ini terdapat data yang outliers sehingga jika menggunakan metrik evaluasi **MSE** akan menghasilkan nilai error yang besar sehingga sulit melakukan prediksi maka dari itu digunakan **RMSE** yang merupakan kuadrat dari nilai **MSE** sehingga menghasilkan nilai yang lebih intuitif dan mudah dilakukan intepretasi. Jika pada **RMSE** kita mendapatkan kesalahan dengan memperhitungkan outliers maka pada **MAE** kita dapat mengintepretasikan kesalahan absolut yang terjadi. **MAPE** merupakan metrik evaluasi yang memiliki satuan persen dengan arti semakin mendekati nilai 0 maka hasil semakin baik. **MAPE** sangat berguna untuk mengintepretasikan hasil error dalam bentuk persen sehingga dapat dengan mudah untuk memahaminya.

**- Root Mean Square Error (RMSE)** adalah salah satu metrik evaluasi yang umum digunakan dalam machine learning, khususnya dalam konteks regresi. Metrik ini digunakan untuk mengukur sejauh mana model regresi cocok dengan data aktual atau sejauh mana perbedaan antara nilai yang diprediksi oleh model dengan nilai sebenarnya.

Berikut adalah deskripsi mengenai RMSE:

1. **Mengukur Kesalahan**: RMSE mengukur kesalahan model regresi dengan menghitung akar kuadrat rata-rata dari perbedaan antara nilai yang diprediksi oleh model dengan nilai sebenarnya (ground truth). Ini berarti bahwa RMSE memberikan bobot lebih besar pada kesalahan besar, yang berarti kesalahan besar akan lebih berdampak pada nilai RMSE daripada kesalahan kecil.

2. **Satuan yang Sama**: RMSE memiliki satuan yang sama dengan target variabel, yang membuatnya mudah diinterpretasikan. Sebagai contoh, jika Anda memodelkan harga apartemen, RMSE akan memiliki satuan mata uang yang sama dengan harga apartemen, seperti dolar atau won.

3. **Proses Perhitungan**: Untuk menghitung RMSE, langkah-langkah berikut dapat diikuti:
   - Selisih antara nilai yang diprediksi oleh model (ŷ) dan nilai sebenarnya (y) dihitung untuk setiap data poin.
   - Kuadrat dari selisih ini dihitung.
   - Rata-rata dari semua kuadrat selisih ini diambil.
   - Akar kuadrat dari rata-rata kuadrat selisih dihitung untuk mendapatkan nilai RMSE.  
   ![image-1.png](RMSE_FORMULA.png)  
    Matematisnya, RMSE dapat dihitung seperti gambar formula diatas:
    di mana (n) adalah jumlah data poin, (yj) adalah nilai sebenarnya, dan (ŷj) adalah nilai yang diprediksi oleh model untuk data poin ke-(j).

4. **Interpretasi**: Semakin kecil nilai RMSE, semakin baik model Anda dalam memprediksi data sebenarnya. Sebaliknya, semakin besar nilai RMSE, semakin besar kesalahan prediksi model Anda.

5. **Perbandingan**: RMSE sering digunakan untuk membandingkan kinerja beberapa model regresi berbeda. Model dengan RMSE yang lebih rendah dianggap lebih baik dalam melakukan prediksi.

**- Mean Absolute Error (MAE)** adalah salah satu metrik evaluasi yang digunakan dalam statistik dan machine learning untuk mengukur sejauh mana model prediksi berbeda dari nilai sebenarnya. MAE mengukur kesalahan prediksi dengan mengambil rata-rata dari selisih absolut antara prediksi model dan nilai sebenarnya. Semakin rendah nilai MAE, semakin baik kinerja model dalam melakukan prediksi.
Berikut adalah deskripsi terkait MAE:
1. **Mengukur Kesalahan**: MAE mengukur kesalahan model dengan menghitung rata-rata dari selisih absolut antara nilai yang diprediksi oleh model dan nilai sebenarnya.
2. **Bobot yang Sama**: MAE memberikan bobot yang sama pada semua kesalahan, tanpa memperhatikan apakah kesalahan tersebut besar atau kecil.
3. **Proses Perhitungan**: Perhitungan MAE biasanya melibatkan langkah-langkah berikut:
   - Selisih absolut antara nilai yang diprediksi oleh model (ŷ) dan nilai sebenarnya (y) dihitung untuk setiap data poin.
   - Rata-rata dari semua selisih ini diambil untuk mendapatkan nilai MAE.  
   ![image-2.png](MAE_FORMULA.png)   
   Matematisnya, MAE dapat dihitung dengan formula diatas:
   MAE = (1/n) * Σ |yj - ŷj|
   di mana (n) adalah jumlah data poin, (yj) adalah nilai sebenarnya, dan (ŷj) adalah nilai yang diprediksi oleh model untuk data poin ke-(j).

4. **Interpretasi**: Semakin kecil nilai MAE, semakin baik model Anda dalam memprediksi data sebenarnya. MAE memberikan ukuran kesalahan rata-rata dalam satuan yang sama dengan target variabel, sehingga mudah diinterpretasikan.
***
**- MAPE (Mean Absolute Percentage Error)** adalah metrik evaluasi yang digunakan untuk mengukur sejauh mana model prediksi berbeda dari nilai sebenarnya dalam bentuk persentase. MAPE sangat berguna dalam kasus ingin mengukur akurasi prediksi dalam konteks persentase kesalahan relatif terhadap nilai sebenarnya. Metrik ini umumnya digunakan dalam peramalan dan perencanaan untuk mengukur kualitas model prediksi dalam hal persentase kesalahan.
Berikut adalah deskripsi terkait MAPE:
1. **Mengukur Kesalahan**: MAPE adalah metrik yang mengukur kesalahan model dalam bentuk persentase dari kesalahan relatif terhadap nilai sebenarnya.
2. **Satuan**: MAPE berguna ketika Anda ingin memahami sejauh mana model regresi menghasilkan kesalahan dalam bentuk persentase dari nilai sebenarnya.
3. **Proses Perhitungan**: Perhitungan MAPE melibatkan langkah-langkah berikut:
   - Selisih absolut antara nilai yang diprediksi oleh model (ŷ) dan nilai sebenarnya (y) dihitung untuk setiap data poin.
   - Selisih absolut ini dibagi oleh nilai sebenarnya dan kemudian diambil rata-rata untuk menghasilkan persentase kesalahan.
   - Nilai persentase kesalahan ini bisa diubah menjadi persen dengan mengalikan dengan 100.  
   ![image-3.png](MAPE_FORMULA.png) 
   Matematisnya, MAPE dapat dihitung dengan formula diatas:
   MAPE = (1/n) * Σ (|yj - ŷj| / |yj|) * 100
   di mana (n) adalah jumlah data poin, (yj) adalah nilai sebenarnya, dan (ŷj) adalah nilai yang diprediksi oleh model untuk data poin ke-(j).

4. **Interpretasi**: Semakin kecil nilai MAPE, semakin baik model Anda dalam memprediksi data sebenarnya dalam bentuk persentase kesalahan relatif.

### VIF (Variance Inflation Factor)

Visualisasi dari correlation matrix diatas menggambarkan terdapat korelasi/hubungan dari beberapa variabel independen yang dapat menimbulkan **multicollinearity** yang dapat mempengaruhi pemodelam dalam regresi. Maka dari itu kita akan melakukan pengecekan nilai **VIF** untuk melihat sejauh mana hubungan/korelasi yang terjadi antar variabel independen

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["Variables"] = X.columns
    vif["Nilai VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
calc_vif(df[['N_FacilitiesNearBy(ETC)', 'N_FacilitiesNearBy(PublicOffice)', 'N_SchoolNearBy(University)', 'N_Parkinglot(Basement)', 'YearBuilt', 'N_FacilitiesInApt', 'Size(sqf)']])
```

Nilai VIF yang dihasilkan dari beberapa variabel memiliki **nilai VIF diatas 10** yang memiliki arti adanya multicollinearity yang signifikan antar variabel independen sehingga nantinya sulit membedakan pengaruh masing - masing variabel terhadap variabel independen (fitur).
Dikarenakan hal tersebut maka pemodelan **linear regression kurang baik** digunakan karena terpengaruh terhadap **multicollinearity** maka dari itu kita akan menghindarkan model linear regression dalam **benchmarking model**

### Handling Outliers

Terdapat variables yang memiliki nilai outliers yaitu kolom **`Size(sqf)` dan `SalePrice`** maka dari itu kita harus melakukan handling pada kolom tersebut

```python
# Membuat fungsi untuk mengecek outliers
def find_anomalies(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit = iqr*1.5
    print(f'IQR: {iqr}')

    limit_bawah = q1 - limit
    limit_atas = q3 + limit
    print(f'limit_bawah: {limit_bawah}')
    print(f'limit_atas: {limit_atas}')
```

Metode yang digunakan dalam pengecekan outliers dilakukan berdasarkan hasil analisis statistik serta visualisasi histogram terkait distribusi data bahwa distribusi dari **data tidak normal** maka untuk melakukan pengecekan outliers lebih cocok menggunakan metode **IQR (Interquartil Range)**

## Data Preprocessing

Data preprocessing adalah tahap penting dalam pemodelan data yang bertujuan untuk menyiapkan data sebelum data tersebut digunakan untuk melatih model. Proses ini membantu meningkatkan kualitas data, mengurangi noise, dan membuat data lebih sesuai untuk digunakan dalam algoritma machine learning. Berikut merupakan beberapa langkah yang dilakukan dalam tahapan data preprocessing:

**1. Transformasi Data**:
   - Normalisasi data: Mengubah skala data sehingga nilainya memiliki rentang yang serupa dalam hal ini scaling yang digunakan adalah **Robust Scaler** karena kita memiliki variabel yang **outliers**
   - Encoding kategori: Mengubah variabel kategori menjadi bentuk numerik yang dapat digunakan oleh model, dalam proses ini kita akan melakukan **One Hot Encoding** dan **Ordinal Encoding**
   - Transformasi fitur: Menerapkan fungsi matematika pada fitur agar sesuai dengan asumsi model  
   
**2. Pemisahan Data (Data Splitting)**:
   - Memisahkan dataset menjadi bagian latihan (training) dan validasi (validation) untuk mengukur performa model secara objektif


```python
# Import library untuk data preprocessing

import category_encoders as ce
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler

from sklearn.model_selection import train_test_split
```

Dari ketiga data kategorikal diatas `HallwayType`, `TimeToSubway`, dan `SubwayStation` memiliki karakteristik yang berbeda - beda sehingga kita perlu menentukan jenis encoding yang cocok dari tiap kategorikal data tersebut :

- `HallwayType` = berdasarkan data understanding kita bisa mengetahui bahwa `HallwayType` merupakan jenis atau tipe dari apartemen yang terbagi menjadi terraced, mixed, ataupun corridor ketiga jenis ini bukan merupakan urutan sehingga encoding dilakukan dengan **One Hot Encoding**

- `TimeToSubway` = merupakan waktu yang dibutuhkan untuk menuju ke stasiun terdekat maka dari itu terdapat urutan dalam data kategorikal tersebut sehingga kita harus melakukan encoding dengan menggunakan **Ordinal Encoding**

- SubwayStation = nama stasiun yang berarti tidak ada perbedaan antara unique tersebut(sejajar) dan jumlah unique dari fitur tersebut hanya 8 maka kita dapat melakukan metode **One Hot Encoding**

#### Ordinal Encoding `TimeToSubway`

```python
# Ordinal Mapping dan Encoding

ordinal_encoding = ce.OrdinalEncoder()
ordinal_var = ['TimeToSubway']
ordinal_mapping = [
    {'col':'TimeToSubway', 'mapping':{'no_bus_stop_nearby':0, '15min~20min':1, '10min~15min':2, '5min~10min':3, '0-5min':4}}
]
```

### Benchmarking Model

Setelah melakukan tahapan data preprocessing maka kita sudah memiliki data yang siap untuk dilakukan modeling. Terdapat beberapa model yang akan kita gunakan sebagai benchmarking model untuk mendapatkan perbandingan kinerja dari tiap modelnya dan dapat melakukan pemilihan model yang terbaik.

Berikut merupakan beberapa model yang akan kita gunakan dalam proses ini:

**1. Ridge**:

Ridge adalah metode regresi linier yang juga digunakan untuk mengurangi overfitting.
Model ini menambahkan regularisasi L2 pada fungsi biaya dengan menerapkan penalti pada koefisien yang besar.
Regularisasi L2 membantu mencegah koefisien menjadi terlalu besar dan menghasilkan model yang lebih stabil.

**2. Lasso (Least Absolute Shrinkage and Selection Operator)**:

Lasso adalah metode regresi linier yang digunakan untuk mengurangi overfitting.
Model ini menambahkan regularisasi L1 pada fungsi biaya dengan menerapkan penalti pada koefisien yang tidak signifikan, sehingga beberapa koefisien bisa menjadi nol.
Ini membantu dalam pemilihan fitur (feature selection) dan menghasilkan model yang lebih sederhana.

**3. KNN Regressor (K-Nearest Neighbors Regressor)**:

KNN Regressor adalah model regresi non-parametrik yang menghitung nilai target berdasarkan rata-rata dari k-nearest neighbors (tetangga terdekat) dalam data latih.
Nilai k adalah jumlah tetangga terdekat yang digunakan untuk menghitung nilai prediksi.

**4. Decision Tree Regressor**:

Decision Tree Regressor adalah model yang menggunakan struktur pohon untuk memprediksi nilai target berdasarkan aturan pemisahan berdasarkan fitur-fitur data.
Masing-masing node dalam pohon memilih fitur terbaik untuk memisahkan data, dan daun pohon berisi prediksi nilai regresi.

**5. RandomForest Regressor**:

RandomForest Regressor adalah ensemble model yang terdiri dari banyak Decision Trees.
Ini menggabungkan prediksi dari banyak pohon keputusan (Decision Trees) untuk menghasilkan prediksi yang lebih stabil dan akurat.

**6. Gradient Boosting Regressor**:

Gradient Boosting Regressor adalah ensemble model yang menggabungkan banyak model regresi (biasanya Decision Trees) secara berurutan.
Setiap model dibuat untuk mengoreksi kesalahan prediksi model sebelumnya, sehingga secara bertahap meningkatkan kinerja model.

**7. XGBoost Regressor**:

XGBoost (Extreme Gradient Boosting) Regressor adalah varian dari Gradient Boosting yang sangat efisien dan efektif.
Ini menggunakan teknik optimasi dan regularisasi khusus untuk meningkatkan akurasi prediksi dan menghindari overfitting.

```python
# Definisi algoritma yang digunakan
rd = Ridge(alpha=1)
ls = Lasso(alpha=1)
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=1)
rf = RandomForestRegressor(random_state=1)
gb = GradientBoostingRegressor(random_state=1)
xgb = XGBRegressor(random_state=1)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_rd = TransformedTargetRegressor(rd, func=np.log, inverse_func=np.exp)
log_ls = TransformedTargetRegressor(ls, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_gb = TransformedTargetRegressor(gb, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)

# Scaling data
scaler = RobustScaler()

# Kandidat algoritma
models = [log_rd, log_ls, log_knn, log_dt, log_rf, log_gb, log_xgb]

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

# Mencari algoritma terbaik berdasarkan nilai MAE
for i in models:
    
    crossval = KFold(n_splits=5, shuffle=True, random_state=1)

    estimator = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', i)
    ])

    # RMSE
    model_cv_rmse = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_root_mean_squared_error', 
        error_score='raise'
        )

    print(model_cv_rmse, i)

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE
    model_cv_mae = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_error', 
        error_score='raise'
        )

    print(model_cv_mae, i)

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_percentage_error', 
        error_score='raise'
        )

    print(model_cv_mape, i)

    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())
```

### Hyperparameter Tuning Gradient Boosting

```python
# Kedalaman pohon
max_depth = list(np.arange(1,6))

# learning rate
learning_rate = list(np.arange(1,15)/100)

# Jumlah pohon
n_estimators = list(np.arange(100,301))

# Jumlah sample sebelum splitting
min_samples_split = list(np.arange(2, 10))

# Jumlah sample sesudah splitting
min_samples_leaf = list(np.arange(1, 20))

# Jumlah fitur yang digunakan untuk splitting (% of total number of train set columns)
max_features = list(np.arange(1,6))
```

```python
# Model 1 dengan hyperparameter tuning
gb = GradientBoostingRegressor(random_state=1)

# Membuat algorithm chains
gb_pipeline = Pipeline([
    ('preprocessing', transformer),
    ('scaler', scaler),
    ('model', gb)
])

# Hyperparameter space Gradient Boost
hyperparam_space_gb = {
    'model__max_depth':max_depth ,
    'model__learning_rate':learning_rate,
    'model__min_samples_split':min_samples_split,
    'model__min_samples_leaf':min_samples_leaf,
    'model__n_estimators':n_estimators,
    'model__max_features':max_features
}

crossval = KFold(n_splits=5, shuffle=True, random_state=0)

# Hyperparameter tuning
random_gb = RandomizedSearchCV(
    gb_pipeline, 
    param_distributions = hyperparam_space_gb,
    n_iter = 50,
    cv = crossval, 
    scoring = 'neg_mean_absolute_error', 
    n_jobs = -1,
    random_state = 1  
)
```

```python
#  Fitting data training untuk mencari parameter terbaik
random_gb.fit(X_train, y_train)
```

```python
# Cek skor dan params terbaik
print('Gradient Boosting Regressor')
print('Best_score:', random_gb.best_score_)
print('Best_params:', random_gb.best_params_)
```

### Residu Analysis

Analisis residu adalah salah satu tahap penting dalam evaluasi model regresi atau model statistik lainnya. Residu adalah selisih antara nilai sebenarnya (data aktual) dan nilai yang diprediksi oleh model. Sebelumnya kita telah mendapatkan hasil final model yang akan kita gunakan yaitu **XGB** maka dari itu selanjutnya kita akan melakukan pengecekan terkait dengan residu dari pemodelan tersebut. **Tujuan** dari analisis residu adalah **memeriksa sejauh mana model regresi sesuai dengan data aktual** ataupun **melihat pola dari residu yang dihasilkan**

```python
# Dataframe berisi data actual, prediksi, dan residu
final = pd.DataFrame({
    "Test" : y_test,
    "Predict" : y_pred_xgb_tuning
})
final['residu'] = final['Predict'] - final['Test']
final.reset_index(drop=True, inplace=True)

# buat looping label keterangan
ket = []
for i in final['residu']:
    if i >= 46633 :
        ket.append('overestimated')
    elif i <= -46633 :
        ket.append('underestimated')
    else:
        ket.append('reasonable')

final['keterangan'] = pd.DataFrame(ket)


# Menghitung jumlah data underestimated, overestimated, dan reasonable
count_underestimated = (final['keterangan'] == 'underestimated').sum()
count_overestimated = (final['keterangan'] == 'overestimated').sum()
count_reasonable = (final['keterangan'] == 'reasonable').sum()

# Menghitung total jumlah data
total_data = final.shape[0]

# Menghitung persentase underestimated dan overestimated
percentage_underestimated = (count_underestimated / total_data) * 100
percentage_overestimated = (count_overestimated / total_data) * 100

# Menampilkan hasil
print(final.head())
print(f'Persentase Underestimated: {percentage_underestimated:.2f}%')
print(f'Persentase Overestimated: {percentage_overestimated:.2f}%')
```

Berdasarkan hasil final pemodelan dengan **XGB** diperoleh nilai **RMSE** sebesar **46633** yang dapat diintepretasikan bahwa data prediksi `SalePrice` terdapat error yang wajar sebesar **46633**.  
Selanjutnya diperoleh rasio hasil prediksi yang termasuk **overestimated** (hasil prediksi jauh lebih tinggi dari aktual) sebesar **18.85%** dan rasio **underestimated** (hasil prediksi jauh lebih rendah dari aktual) sebesar **12.01%**

**Cara kerja SHAP** adalah sebagai berikut:

- Memodelkan semua kombinasi fitur: SHAP menghitung kontribusi fitur dengan mempertimbangkan semua kemungkinan kombinasi fitur dalam data, dan kemudian membandingkannya dengan kontribusi fitur saat fitur tersebut dihilangkan dari prediksi.

### Model Limitation

Setiap model memiliki keterbatasan tertentu yang dapat memengaruhi kinerja dan kemampuan model tersebut. Perlunya membuat model limitation untuk suatu model agar dapat memahami keterbatasan model dan membantu memiliki pandangan yang lebih realistis tentang apa yang dapat dicapai oleh model tersebut. Ini membantu menghindari ekspektasi yang tidak realistis dan mengarahkan eksperimen atau pengembangan model ke arah yang lebih produktif.

```python

def mean_absolute_percentage_error(y_true, y_pred):
    return 100 * (abs(y_true - y_pred) / y_true).mean()

list_mae = []
list_mape = []
nrow = []

price_ranges = [(0, 100000), (100000, 200000), (200000, 300000), (300000, 400000), (500000, float('inf'))]

for price_range in price_ranges:
    min_price, max_price = price_range
    data = pd.DataFrame({'y_test': y_test, 'y_pred': y_pred_xgb_tuning})
    data = data[(data['y_test'] >= min_price) & (data['y_test'] < max_price)]
    nrow.append(len(data))
    mae = mean_absolute_error(data['y_test'], data['y_pred'])
    mape = mean_absolute_percentage_error(data['y_test'], data['y_pred'])  # Calculate MAPE
    list_mae.append(mae)
    list_mape.append(mape)

result_df = pd.DataFrame({
    'Range Harga': [f'{min_price}-{max_price}' for min_price, max_price in price_ranges],
    'Jumlah Data': nrow,
    'MAE': list_mae,
    'MAPE (%)': list_mape  # Display MAPE as a percentage
})

print(result_df)
```

```python
range_harga = result_df['Range Harga']
mape_percentage = result_df['MAPE (%)']

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(range_harga, mape_percentage, color='skyblue')
plt.xlabel('Range Harga')
plt.ylabel('MAPE (%)')
plt.title('MAPE by Price Range')


plt.axhline(y=25, color='red', linestyle='--', label='Threshold (25%)')

# Show the plot
plt.tight_layout()
plt.legend()
plt.show()
```

Grafik diatas merupakan hasil visualisasi range `SalePrice` dengan nilai **MAPE** disetiap rangenya. Didapatkan hasil bahwa untuk `SalePrice` yang terdapat pada range 0-100000 (WON) memiliki nilai **MAPE** yang melebihi **nilai threshold 25%** yang merupakan batas nilai **MAPE** yang dapat diterima walaupun memiliki tingkat akurasi rendah [MAPE](https://www.vexpower.com/brief/mean-absolute-percentage-error#:~:text=The%20mean%20absolute%20percentage%20error,actual%20values%20and%20your%20prediction.). Kemudian nilai **MAPE** dibawah 15% terdapat pada range harga 200000 - 400000 (WON) pada rentang sekian didapatkan nilai **MAPE** yang memiliki arti nilai prediksi mendekati benar. Kemudian pada range harga lebih dari 500000 menghasilkan nilai **MAPE** dibawah threshold **25%** namun data yang terdapat pada range harga ini hanya 7 sehingga terdapat indikasi bahwa nilai **MAPE** belum dapat merepresentasikan secara keseluruhan pada range harga tersebut.
